```python
def sample_mean_calculator(input_dist, sample_size, n_samples):
    sample_means = []
    for i in range(n_samples):
        sample = np.random.choice(input_dist, size=sample_size, replace=False)
        sample_means.append(sample.mean())
    return sample_means
```

```python
df.isna().sum()
```

```python
y=df['Age']
```

```python
x=df.drop(['Age'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
categorical_columns=[col for col in x_train.columns if x_train[col].dtype=='object']
numerical_columns=[col for col in x_train.columns if col not in categorical_columns]
```

```python
numeric_transformer=StandardScaler()
categorical_transformer=OneHotEncoder(handle_unknown='ignore')
```

```python
preprocessor=ColumnTransformer(transformers=[('num',numeric_transformer,numerical_columns),('cat',categorical_transformer,categorical_columns)])
```

```python
#create a linear regression model
model=LinearRegression()
```

```python
#create the pipeline
pipeline=Pipeline(steps=[('preprocessor',preprocessor),('model',model)])
```

```python
#fit the model on the training data
pipeline.fit(x_train,y_train)
```

```python
#make predictions on the test set
y_pred=pipeline.predict(x_test)
```

```python
#evaluate the model performance
mse=mean_squared_error(y_test,y_pred)
print(f'Mean Squared Error:{mse}')
```

```python
#check the prediction
new_data_predictions=pipeline.predict(df)
print(f'predictions for the new data:{new_data_predictions}')
```

```python
#plot the prediction
plt.plot(new_data_predictions)
plt.show()
```
